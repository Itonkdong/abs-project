# Test Experiments

## Experiment 1

```python
%cd src/experiments/viktor/
```

```python
exp_config = ExperimentConfig.get_from_yaml()
task = VmasTask.GIVE_WAY.get_from_yaml()
alg = MappoConfig.get_from_yaml()
model = MlpConfig.get_from_yaml()
critic_model = MlpConfig.get_from_yaml()
```

```python
task.config["max_steps"] = 200
exp_config.train_device = "cpu"
exp_config.max_n_frames = 12_000
```

```python
task.config["final_reward"] = 0.01
task.config["pos_shaping_factor"] = 1.0
task.config["shared_rew"] = True
```

```python
task.config
```

```python
experiment = Experiment(
    task=task,
    algorithm_config=alg,
    model_config=model,
    critic_model_config=critic_model,
    seed=0,
    config=exp_config
)
```

```python
experiment.run()
```
